```python
import pandas as pd
import numpy as np

movies_dataset = pd.read_csv("..\data\movies_metadata_credits_joined2.csv", sep = ",")
```

```python
# Compte le nombre de films
nombre_de_films = 0
for i in movies_dataset["original_title"]:
    if i != "":
        nombre_de_films += 1

print(nombre_de_films)
```

```python
duree_totale = movies_dataset["runtime"].sum()
print(duree_totale)
```

```python
# Dur√©e moyenne des films (en minutes)
duree_moyenne = duree_totale/nombre_de_films
print(duree_moyenne)
```

```python
# Nombre d'acteurs par film en moyenne
compteur = 0
for i in movies_dataset["cast_names"]:
    acteurs_liste = str(i).split(sep = ', ')
    # Compte le nombre total d'acteurs
    for j in acteurs_liste:
        compteur += 1

acteurs_par_film = compteur/nombre_de_films

print(acteurs_par_film)
```
