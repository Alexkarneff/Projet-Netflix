```python
import pandas as pd
import numpy as np

movies_dataset = pd.read_csv("..\data\movies_metadata_credits_joined.csv", sep = ",")
```

```python
# Compte le nombre de films
nombre_de_films = 0
for i in movies_dataset["original_title"]:
    if i != "":
        nombre_de_films += 1

print(nombre_de_films)
```

```python
duree_totale = movies_dataset["runtime"].sum()
print(duree_totale)
```

```python
duree_moyenne = duree_totale/nombre_de_films
print(duree_moyenne)
```
