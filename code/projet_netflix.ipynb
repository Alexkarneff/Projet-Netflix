```python
"""
Les colonnes qui m'ont semblées les plus pertinantes : 
'id', 'original_title', 'genres', 'original_language' et 'runtime' sont explicitement demandées dans l'énoncé.
'release_date', 'production_countries', 'production_companies', 'status' sont des infos suplémentaires que j'ai trouvé intéressantes.
'popularity' peut être intéressante pour les recommandations. 
"""


movie_metadata_filtre = movie_metadata[["id", "original_title", "genres", "release_date", "original_language", "production_countries", "production_companies", "runtime", "popularity", "status"]]
movie_metadata_filtre
```

```python
"""
J'ai suprimmé les lignes avec des données manquantes, autre possibilité : remplacer les données manquantes.
Il manquait des données dans toutes les colonnes sauf 'id', 'original_title' et 'genres'.
"""

movie_metadata_filtre.dropna(inplace=True)
movie_metadata_filtre
```

```python
movie_metadata_filtre.info()
```

```python
"""
89 langues différentes. À voir comment une recherche par langue peut être implémentée.
"""
print(movie_metadata_filtre["original_language"].value_counts().iloc[0:50:])
print(movie_metadata_filtre["original_language"].value_counts().iloc[51:89:])
```

```python
"""
J'ai converti les données de la colonne 'popularity' en float pour pouvoir les comparer.
Conversion des id en int64 pour qu'ils aient le même type que dans la table credits.
"""
movie_metadata_filtre["popularity"] = pd.to_numeric(movie_metadata_filtre["popularity"])
movie_metadata_filtre['id'] = pd.to_numeric(movie_metadata_filtre['id'])
movie_metadata_filtre.info()
```

```python
movie_metadata_filtre.sort_values("popularity", ascending=True)
```
