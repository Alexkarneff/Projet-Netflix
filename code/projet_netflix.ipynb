```python
"""
Les colonnes qui m'ont semblées les plus pertinantes : 
'id', 'original_title', 'genres', 'original_language' et 'runtime' sont explicitement demandées dans l'énoncé.
'release_date', 'production_countries', 'production_companies', 'status' et 'overview' sont des infos suplémentaires que j'ai trouvé intéressantes.
'popularity' peut être intéressante pour les recommandations. 
"""


movie_metadata_filtre = movie_metadata[["id", "original_title", "genres", "release_date", 
                                        "original_language", "production_countries", "production_companies", 
                                        "runtime", "popularity", "status", "overview"]]
movie_metadata_filtre.head()
```

```python
"""
Est-ce utile de suprimer les lignes dans lesquelles il manque des infos?
Oui. C'est triste mais c'est comme ça. 
"""

movie_metadata_filtre.dropna(inplace=True)

movie_metadata_filtre
```

```python
movie_metadata_filtre.info()
```

```python
"""
89 langues différentes. À voir comment une recherche par langue peut être implémentée.
"""
print(movie_metadata_filtre["original_language"].value_counts().iloc[0:50:])
print(movie_metadata_filtre["original_language"].value_counts().iloc[51:89:])
```

```python
"""
J'ai converti les données de la colonne 'popularity' en float pour pouvoir les comparer.
Conversion des id en int64 pour qu'ils aient le même type que dans la table credits.
"""
movie_metadata_filtre["popularity"] = pd.to_numeric(movie_metadata_filtre["popularity"])
movie_metadata_filtre['id'] = pd.to_numeric(movie_metadata_filtre['id'])

movie_metadata_filtre.info()
```

```python
movie_metadata_filtre.sort_values("popularity", ascending=False)
```

```python
movies_credits_joined = pd.merge(movie_metadata_filtre, movie_credits, on='id')

movies_credits_joined.info()
```

```python
movies_credits_joined.info()
```

```python
print(movies_credits_joined['genres'])
```

```python
import pandas as pd

import ast

movies_credits_joined['genres'] = movies_credits_joined['genres'].apply(ast.literal_eval)

movies_credits_joined['genres'] = movies_credits_joined['genres'].apply(lambda x: ', '.join([d['name'] for d in x]))

movies_credits_joined['genres']
```

```python
movies_credits_joined['cast'] = movies_credits_joined['cast'].apply(ast.literal_eval)

movies_credits_joined['cast'] = movies_credits_joined['cast'].apply(lambda x: ', '.join([d['name'] for d in x]))
movies_credits_joined['cast']
```

```python
movies_credits_joined['production_countries'] = movies_credits_joined['production_countries'].apply(ast.literal_eval)

movies_credits_joined['production_countries'] = movies_credits_joined['production_countries'].apply(lambda x: ', '.join([d['name'] for d in x]))
movies_credits_joined['production_countries']
```

```python
movies_credits_joined.info()
```

```python
movies_credits_joined.to_csv('movies_metadata_credits_joined.csv', index=False)
```
